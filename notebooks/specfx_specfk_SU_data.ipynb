# Fourier transform of SU data

## Read a SU file

```python
# Import the SUdata class from nessi.io module
from nessi.io import SUdata
import numpy as np

# Declare
dobs = SUdata()

# Read the SU file
dobs.read('data/musc_F50_01.su')

dobsfx = dobs.specfx()
dobsfk = dobs.specfk()
```

## Visualize data

Since many useful informations are stored in the SU header, plotting data is easy and the function's parameters are almost all related to the output image layout.  

```python
# Import matplotlib
import matplotlib.pyplot as plt

# Show data
fig = plt.figure(figsize=(12,5))
plt.subplot(121)
dobsfx.image(label2='trace number', label1='frequency [Hz]', title='Visualize FX', legend=1, cmap='jet')
plt.subplot(122)
dobsfk.image(label2='K', label1='frequency [Hz]', title='Visualize FK', legend=1, cmap='jet')
plt.show()
```
