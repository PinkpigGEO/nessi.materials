# SU manipulation

```python
import numpy as np
import matplotlib.pyplot as plt

from nessi.io import SUdata
```

```python
musc = SUdata()

musc.read('musc_F50_01.su')
muscw = musc.wind(tmin=0.0, tmax=0.25)
musct = muscw.taper(tr1=40, tr2=40, tbeg=100, tend=100)

fig = plt.figure(figsize=(15,4))
plt.subplot(131)
musc.image(clip=0.05, legend=1)
plt.subplot(132)
muscw.image(clip=0.05, legend=1)
plt.subplot(133)
musct.image(clip=0.05, legend=1);
```
